# API Call

````python
from openai import OpenAI
from dotenv import load_dotenv

from concurrent.futures import ThreadPoolExecutor
import threading

jsonlock = threading.Lock()


load_dotenv()
api_key = os.getenv('API_KEY')

if api_key is None:
    print("API_KEY not found in .env file")
````

````python
def save_file(file_name, file_content):
    """Save the file content to a new file"""
    with open(file_name, 'a') as f:
        f.write(file_content + '\n')

def save_error(file_content):
    """Save the error to a new file"""
    with open('error.txt', 'a') as f:
        f.write(file_content + '\n')

def show_output(response):
    print(response.choices[0].message.content)
    save_file('output.txt', response.choices[0].message.content)


def show_streaming_output(response, pasal):
    # Menangani respons streaming
    collected_chunks = []
    collected_messages = []

    for chunk in response:
        if chunk.choices[0].delta.content is not None:
            print(chunk.choices[0].delta.content, end="", flush=True)
        collected_chunks.append(chunk)
        collected_messages.append(chunk.choices[0].delta.content)

    full_reply = ''.join([m for m in collected_messages if m is not None])

    file_name = 'output_' + str(pasal) + '.txt'
    save_file(file_name, full_reply)
    return full_reply
````

````python
def add_to_json(full_reply, firstCQA):
    with jsonlock:
        jsonstart = full_reply.find("[")
        jsonend = full_reply.rfind("]") + 1
        replyJson = json.loads(full_reply[jsonstart:jsonend])
        data = []
        with open(os.path.join(datasetDir, "datasetVariance.json"), "r") as f:
            if os.stat(f.name).st_size == 0:
                data = []
            else:
                data = json.load(f)

        with open(os.path.join(datasetDir, "datasetVariance.json"), "w") as f:
            for i in replyJson:
                data.append(i)
            data.append(firstCQA)
            json.dump(data, f, indent=4)


def call_openai_api(pasal, data, ke):
    try:
        client = OpenAI(api_key=api_key, base_url="https://api.deepseek.com")

        response = client.chat.completions.create(
        model="deepseek-chat",
        # model='deepseek-reasoner',
        messages=[
            {"role": "system", "content": "Kamu adala asisten yang paham tentang uu no 34 tahun 2004 tentang TNI dan kamu akann membantu saya untuk memperkaya CQA yang saya kirim menjadi 3 variasi, kembalikan dalam bentuk json"},
            {"role": "user", "content": f"""
                Saya akan mengirimkan Satu CQA yang sudah ada, tolong buatkan 3 variasi baik berupa bahasa cara bertanya dengan lebih tidak formal, baku, dan variasi lainnya untuk Question, sedangkan untuk Answer tetap gunakan jawaban formal dan bervariasi dari yang saya kirim
                 - Buat answer yang panjang minimal 7 kalimat
                 - gunakan format Markdown seperti (#, ##, ###), numbering, bullet, dan lain-lain
                CQA yang sudah ada adalah:
                {data}
                Ini adalah pasal yang terkait :
                {jsonUU[pasal -1]}

                Buat CQA baru dalam format json seperti yang saya contohkan.
                Format json yang saya inginkan adalah sebagai berikut:
                ```json
                [
                    {{
                        "Context": "",
                        "Question": "",
                        "Answer": ""
                    }},
                    {{
                        "Context": "",
                        "Question": "",
                        "Answer": ""
                    }},
                    {{
                        "Context": "",
                        "Question": "",
                        "Answer": ""
                    }}
                ]
                ```
                Hanya response dengan json saja, tanpa penjelasan atau tambahan lainnya.
                
            """},
            ],
            stream=True
        )

        # show_output()
        full_reply = show_streaming_output(response=response, pasal=pasal)

        add_to_json(full_reply, data)
    except Exception as e:
        print("Error: ", e)
        print("Pasal: ", pasal)
        print("Data: ", data)
        save_error(f"Pasal: {pasal}, Data ke: {ke}, Error: {e}")
        pass
````

````python
def process_pasal_data(pasal, data):
    for i in range(len(data)):
        ke = i + 1
        print("Pasal: ", pasal, "Data ke: ", ke)
        call_openai_api(pasal, data[i], ke)

with ThreadPoolExecutor(max_workers=26) as executor:
    for i in range(len(CQAJson)):
        pasal = int(CQAJson[i]['pasal'])
        data = CQAJson[i]['data']
        if pasal > 52:
            continue
        executor.submit(process_pasal_data, pasal, data)
````
