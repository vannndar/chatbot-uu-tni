```python
import glob
import json
import pandas as pd

# List all JSON files in the folder
file_paths = glob.glob('/content/dataset_1-26/*.json')

records = []

for fp in file_paths:
    with open(fp, 'r') as f:
        data = json.load(f)
    filename = fp.split('/')[-1]

    # Handle both list and dict JSON structures
    if isinstance(data, list):
        for item in data:
            item['filename'] = filename
            records.append(item)
    elif isinstance(data, dict):
        data_entry = data.copy()
        data_entry['filename'] = filename
        records.append(data_entry)
    else:
        # For unexpected types, store raw data
        records.append({'filename': filename, 'data': data})

# Create DataFrame
df = pd.DataFrame(records)
df.head()
```
