# API Call

````python
from openai import OpenAI
from dotenv import load_dotenv

load_dotenv()
api_key = os.getenv('API_KEY')
````

````python
def save_file(file_name, file_content):
    """Save the file content to a new file"""
    with open(file_name, 'wb') as f:
        if isinstance(file_content, str):
            f.write(file_content.encode('utf-8'))
        else:
            f.write(file_content)


def show_output(response):
    print(response.choices[0].message.content)
    save_file('output.txt', response.choices[0].message.content)


def show_streaming_output(response, pasal):
    # Menangani respons streaming
    collected_chunks = []
    collected_messages = []

    for chunk in response:
        if chunk.choices[0].delta.content is not None:
            print(chunk.choices[0].delta.content, end="", flush=True)
        collected_chunks.append(chunk)
        collected_messages.append(chunk.choices[0].delta.content)

    full_reply = ''.join([m for m in collected_messages if m is not None])

    save_file('output_{pasal}.txt', full_reply)

    return full_reply
````

````python
def add_to_json(full_reply, firstCQA):
    jsonstart = full_reply.find("[")
    jsonend = full_reply.rfind("]") + 1
    replyJson = json.loads(full_reply[jsonstart:jsonend])
    data = []
    with open(os.path.join(path, "datasetVariance.json"), "r") as f:
        if os.stat(f.name).st_size == 0:
            data = []
        else:
            data = json.load(f)

    with open(os.path.join(path, "datasetVariance.json"), "w") as f:
        for i in replyJson:
            data.append(i)
        data.append(firstCQA)
        json.dump(data, f, indent=4)
        

def call_openai_api(pasal, data):
    client = OpenAI(api_key=api_key, base_url="https://api.deepseek.com")

    response = client.chat.completions.create(
    model="deepseek-chat",
    # model='deepseek-reasoner',
    messages=[
        {"role": "system", "content": "Kamu adala asisten yang paham tentang uu no 34 tahun 2004 tentang TNI dan kamu akann membantu saya untuk memperkaya CQA yang saya kirim menjadi 3 variasi, kembalikan dalam bentuk json"},
        {"role": "user", "content": f"""
            Saya akan mengirimkan Satu CQA yang sudah ada, tolong buatkan 3 variasi baik berupa bahasa cara bertanya dengan lebih tidak formal
            CQA yang sudah ada adalah:
            {data}
            Ini adalah pasal yang terkait :
            {jsonUU[pasal -1]}

            Buat CQA baru dalam format json seperti yang saya contohkan.
            Format json yang saya inginkan adalah sebagai berikut:
            ```json
            [
                {{
                    "Context": "",
                    "Question": "",
                    "Answer": ""
                }},
                {{
                    "Context": "",
                    "Question": "",
                    "Answer": ""
                }},
                {{
                    "Context": "",
                    "Question": "",
                    "Answer": ""
                }}
            ]
            ```
            Hanya response dengan json saja, tanpa penjelasan atau tambahan lainnya.
            
        """},
        ],
        stream=True
    )

    # show_output()
    full_reply = show_streaming_output(response=response, pasal=pasal)

    add_to_json(full_reply, data)

````

````python
for i in range(len(CQAJson)):
    pasal = int(CQAJson[i]['pasal'])
    data = CQAJson[i]['data']
    for j in range(len(data)):
        print("Pasal ke-", pasal, "data ke-", j)
        call_openai_api(pasal, data[j])
````
